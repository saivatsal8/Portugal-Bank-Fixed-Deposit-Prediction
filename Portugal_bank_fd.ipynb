### data cleaning

```python
df['contact'] = df['contact'].replace({'cellular': 1, 'telephone': 0})
```

```python
three_categories = ['loan', 'housing', 'default']
df[three_categories] = df[three_categories].replace({'yes': 1, 'no': 0, 'unknown': 0})
```

```python
df.pdays = df.pdays.replace(999, 0) # replace 999 with 0 rest all the values are same
df.pdays.unique()
```

```python
df.previous=df.previous.apply(lambda x:1 if x>0 else 0).astype('uint8') # replace all the values with 0
```

```python
df.previous.unique()
```

```python
df.poutcome=df.poutcome.replace({"success":1,'nonexistent':0,'failure':0}).astype('uint8')
df.poutcome
```

```python
df.emp_var.unique()
```

```python
df['emp_var']=df['emp_var'].apply(lambda x:x*-0.0001 if x>0 else x*1)
df.emp_var.unique()
```

```python
df['emp_var']=df['emp_var']*-1
df.emp_var.unique()
```

```python
df['emp_var']=df['emp_var'].apply(lambda x:-np.log(x) if x<1 else np.log(x))
df.emp_var.unique()
```

```python
df.cons_price=df.cons_price*10
df.cons_price
```

```python
df['cons_confg']=df['cons_confg'] * -1
df.cons_confg
```

```python
df.num_employed.unique()
```

```python
log_cols=['num_employed','cons_confg','cons_price']
for col in log_cols:
    df[col]=np.log2(df[col]) # reducing the values for processing
```

### measuring and comparing

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_predlog)
cm
```

```python
conf_matrix = pd.DataFrame(data=cm,columns=['Predicted:0','Predicted:1'],index=['Actual:0','Actual:1'])
```

```python
plt.figure(figsize = (8,5))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap="YlGnBu", cbar=False)
plt.show()
```

```python
from sklearn.metrics import classification_report
result = classification_report(y_test,y_predlog)

# print the result
print(result)
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
# set the figure size
plt.rcParams['figure.figsize']=(8,5)

fpr, tpr, thresholds = roc_curve(y_test, y_predlog)

# plot the ROC curve
plt.plot(fpr,tpr)

# set limits for x and y axes
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])

# plot the straight line showing worst prediction for the model
plt.plot([0, 1], [0, 1],'r--')

# add the AUC score
plt.text(x = 0.05, y = 0.8, s =('AUC Score:', round(roc_auc_score(y_test, y_predlog),4)))

# name the plot, and both axes
plt.xlabel('False positive rate (1-Specificity)')
plt.ylabel('True positive rate (Sensitivity)')

# plot the grid
plt.grid(True)
plt.show()
```

```python
from sklearn import metrics
cols = ['Model', 'AUC Score', 'Precision Score', 'Recall Score','Accuracy Score','f1-score']
result_table=pd.DataFrame(columns=cols)
values=pd.Series({'Model': "Logistic Regression",
                     'AUC Score' : metrics.roc_auc_score(y_test, y_predlog),
                 'Precision Score': metrics.precision_score(y_test, y_predlog),
                 'Recall Score': metrics.recall_score(y_test, y_predlog),
                 'Accuracy Score': metrics.accuracy_score(y_test, y_predlog),
                  'f1-score':metrics.f1_score(y_test, y_predlog)
    
})
result_table = pd.concat([result_table, values.to_frame().T], ignore_index=True)

result_table
```
